```python
import matplotlib.pyplot as plt
%matplotlib inline
from chainer import functions as F
from chainer import Variable
from chainer.links.model.vision.vgg import prepare
from chainer.links import VGG16Layers
import numpy as np
import time
from PIL import Image
```

```python
model = VGG16Layers()
```

```python
img_ori = Image.open('./lion.png')
img = prepare(img_ori)
img = np.expand_dims(img,0)
```

```python
grad_list = []
noise = 255 * 0.2

for k in range(50):
    r = np.random.randn(img.shape[0],img.shape[1],img.shape[2],img.shape[3]).astype(np.float32)
    img_rand = img + r*noise
    img_rand = Variable(img_rand)
    pred = model(img_rand)
    if k == 0:
        ind = F.argmax(pred['prob']).data
    p = pred['prob'][0,ind]
    p.backward()
    grad_list.append(img_rand.grad)
```

```python
grad = np.mean(np.mean(np.abs(np.array(grad_list)),axis=0)[0],axis=0)
```

```python
img_re = np.array(img_ori.resize([224,224]))
img_grad = np.copy(img_re)
img_grad[:,:,0] *=0
img_grad[:,:,0] = grad / np.max(grad) * 255 * 5 # for visualization
```

```python
plt.figure(figsize=[15,5])
plt.subplot(1,3,1)
plt.imshow(img_re)
plt.title('ori')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(grad)
plt.gray()
plt.title('smoothgrad')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_grad)
plt.title('img + grad')
plt.axis('off')

plt.savefig('smoothgrad.png')
```
